```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('circleresults.csv')

xd = df['xd']
x = df['x']
y = df['y']
yd = df['yd']

plt.plot(xd, label="Xd")
plt.plot(x, label="X")
plt.legend()
plt.xlabel("Step")
plt.ylabel("X Coordinate")
plt.title("Circle Trajectory - Xd vs. X")
plt.show()
```

```python
plt.plot(yd, label="Yd")
plt.plot(y, label="Y")
plt.legend()
plt.xlabel("Step")
plt.ylabel("Y Coordinate")
plt.title("Circle Trajectory - Yd vs. Y")
plt.show()
```
