```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('circleresults.csv')

xd = df['xd']
x = df['x']
y = df['y']
yd = df['yd']

plt.plot(xd,x)
plt.xlabel("Desired X")
plt.ylabel("Robot X")
plt.title("Circle Trajectory - Xd vs. X")
plt.show()
```

```python
plt.plot(yd,y)
plt.xlabel("Desired Y")
plt.ylabel("Robot Y")
plt.title("Circle Trajectory - Yd vs. Y")
plt.show()
```
