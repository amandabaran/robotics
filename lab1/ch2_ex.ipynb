```python
from numpy import *
from matplotlib.pyplot import *
```

### CH.2 Excercise 1

```python
def control(t, y):
#     Control function is the derivative of p(t)
    ux = -4 * cos(t)
    uy = -2 * sin(t)
    return array([ux, uy])

def simulate(Δt, x, u):
    x += Δt * u  #Euler integration
    return x

def sense(x):
#   A perfect sensor where y(t) = p(t)
    return x

# Simulator for circle trajectory:

tf = 2*math.pi #final time
Δt = 0.01 #time step
time = linspace(0, tf, int(tf / Δt) + 1) #time interval


x = array([3.,4.])  #initial position
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t,y)
    x = simulate(Δt, x, u)  #simulate by integration
    x_log.append(copy(x))

x_log = array(x_log)
```

```python
grid()
plot(x_log[:,0], x_log[:,1])
```

### CH.2 Excercise 2

```python
def control(t, y):
#     Modify control function with rotation matrix
    ux = (-2 * sqrt(3) * cos(t) + sin(t))
    uy = (-2 * cos(t)) - (sqrt(3) * sin(t))
    return array([ux, uy])

def simulate(Δt, x, u):
    x += Δt * u  #Euler integration
    return x

def sense(x):
#   A perfect sensor where y(t) = p(t)
    return x

# Simulator for circle trajectory:

tf = 10. #final time
Δt = 0.01 #time step
time = linspace(0, tf, int(tf / Δt) + 1) #time interval


x = array([2.,(sqrt(3)+2.)])  #initial condition 
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t,y)
    x = simulate(Δt, x, u)  #simulate by integration
    x_log.append(copy(x))

x_log = array(x_log)
```

```python
grid()
plot(x_log[:,0], x_log[:,1])
```
